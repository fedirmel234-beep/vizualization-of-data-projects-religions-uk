```python
import pandas as pd
import plotly.express as px
import geopandas as gpd
import matplotlib.pyplot as plt

import numpy as np
```

# Map on the number of people by religions

We decided to show how popular are certain religions in different countries and show the difference between 2010 and 2020 years.

# Data for 2010

```python
df = pd.read_csv("number-of-people-by-religion.csv")
df = df.query("Year == 2010").copy()
religion_col = ["Buddhists", "Hindus", "Muslims", "Jews", "Christians", "No religion", "Other religions"]
for col in religion_col:
    df[col] = (
        df[col]
        .astype(str) #convertation to text
        .str.replace("< 10000", "10000", regex=False) #changing "< 10000" to "10000"
        .str.replace(" ", "", regex=False) #replace spaces
    )
    df[col] = pd.to_numeric(df[col], errors="coerce") #convertation to int, if we cannot do it value converts to Nan
    df = df[df["Code"].fillna("").str.len() == 3].copy()
    #keep values with standatized code f. e. SVK, SWE, etc
    df = df[~df["Code"].fillna("").str.startswith("OWID_")].copy()
    #remove OWID
for religion in religion_col:
    df[f"{religion}_log"] = np.log10(df[religion] + 1)
    fig = px.choropleth(
        df,
        locations="Code",
        color=f"{religion}_log",
        hover_name="Entity",
        hover_data=["Code", religion],
        title=f"{religion} by country (2010, log scale)",
        color_continuous_scale="Blues"
    )
    fig.show()
```
