```python
import json
import pandas as pd
import plotly.express as px
import numpy as np
import geopandas as gpd
```

# Map on the number of people by religions

We decided to show how popular are certain religions in different countries and show the difference between 2010 and 2020 years.

# Data for 2010

```python
with open('data/countries_ukraine_crimea.geojson', encoding='utf-8') as f:
    world_geo = json.load(f)
df = pd.read_csv("data/number-of-people-by-religion.csv")
df = df[
    (df["Year"] == 2010) &
    (df["Code"].fillna("").str.len() == 3) &
    (~df["Code"].fillna("").str.startswith("OWID_"))
        ].copy()
religion_col = ["Buddhists", "Hindus", "Muslims", "Jews", "Christians", "No religion", "Other religions"]

df[religion_col] = (
    df[religion_col]
    .astype(str) #convertation to text
    .replace("< 10000", "10000", regex=False) #changing "< 10000" to "10000"
    .replace(r"\s+", "", regex=True) #replace spaces
)
df[religion_col] = df[religion_col].apply(
    pd.to_numeric,
    errors="coerce"
) #convertation to int, if we cannot do it value converts to Nan
#keep values with standatized code f. e. SVK, SWE, etc
    
    #remove OWID
for religion in religion_col:
    df[f"{religion}_log"] = np.log10(df[religion] + 1)
    fig = px.choropleth(
        df,
        geojson=world_geo,
        locations="Code",
        color=f"{religion}_log",
        featureidkey="properties.ISO3166-1-Alpha-3",
        hover_name="Entity",
        hover_data=["Code", religion],
        title=f"{religion} by country (2010, log scale)",
        color_continuous_scale="Blues"
    )
    fig.show()
```
