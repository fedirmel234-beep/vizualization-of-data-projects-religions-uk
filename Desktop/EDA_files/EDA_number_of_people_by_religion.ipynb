We have a problem in that the columns containing the number of people of a given faith may contain both very small values and “< 10,000” 
The question is what to do with them: 
1) Replace them with something around 10,000 so we can perform calculations with them later, but then there’s a risk that we’ll significantly distort the information
2) Keep the “<10,000” category, but then we can’t perform any calculations with these values
3) Replace all values with categories, sort of smoothing them out, but we can’t do that either, since populations in different countries may vary, and 10,000 people in a country with a population of one million and 40 million need to be coded into different categories

If we solve this problem, we will be able to explore many aspects related to the changing trends in the number of people belonging to a particular religion in a particular country
